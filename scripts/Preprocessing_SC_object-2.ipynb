```python
import pandas as pd
import scanpy as sc
```

```python
filename = "/home/cbb575_jy752/Final_project/Single-Cell_Datasets/From_Anvi/IPDCO_hg_midbrain_UMI.tsv"
#Read in the UMI count matrix
raw_adata = sc.read_text(filename, delimiter='\t', first_column_names=False, dtype='float32')
```

Since the genes ID provided are in ENSG ID format, we will added a new column with the corresponding gene names

```python
#Read in ensgid_gene_symbol conversion file
gene_id_conversion_file = pd.read_csv('/home/cbb575_jy752/palmer_scratch/ensgID_genesymbols/result.txt', sep="\t", names=["ensg_id", "gene_symbol"])
```

```python
gene_id_conversion_file
```

```python
#Merge the genes dataframe with the gene_id_conversion_file keeping the rows in genes dataframe only, adding the gene_symbol column
genes = genes.merge(gene_id_conversion_file, how='left', left_on='gene', right_on='ensg_id')
```

```python
#Count how many genes don't have gene_symbols
genes['gene_symbol'].isna().sum()
```

```python
adata.var_names = genes['gene_symbol']
```

```python
adata.var_names
```

```python
#Drop the genes that didn't map to a gene_symbol
adata = adata[:, adata.var_names.notna()]
#For the genes that are duplicated after mapping, keep only the first record
adata = adata[:, ~adata.var_names.duplicated(keep='first')]
```

```python
adata
```

Now that all the minimum data has been properly formatted and aggregated into an anndata object, the basic per-cell level QC info should be determined.

```python
sc.pp.calculate_qc_metrics(adata, inplace=True)
```

```python
adata.obs
```

Now that the per-cell level QC information has been added, I will add the clinical metadata for the patients on a per-cell basis.

```python
#Read in the aggregated clinical metadata
clinical_metadata = pd.read_csv("/home/cbb575_jy752/Final_project/Single-Cell_Datasets/From_Anvi/Clinical_metadata.csv")
clinical_metadata
```

```python
#Change the PD code in the "Patient ID" column of the clinical metadata to IPD
clinical_metadata['Patient ID'] = clinical_metadata['Patient ID'].str.replace(r'\bIPD(\d+)\b', r'PD\1', regex=True)
clinical_metadata
```

```python
#Drop the column had all missing spaces
clinical_metadata = clinical_metadata.drop("Unnamed: 7", axis=1)
```

```python
clinical_metadata
```

```python
adata.obs['patient'].value_counts()
```

```python
clinical_metadata['Patient ID'].value_counts()
```

```python
#Merge the clinical metadata to the cells in the anndata object observations layer
adata.obs = adata.obs.reset_index().merge(clinical_metadata, left_on='patient', right_on='Patient ID').set_index('barcode')
```

```python
adata.obs
```

Now that the metadata is added to single-cell object, we want to create a UMAP plot that will be used for plotting later

```python
# Perform PCA if not already done
if 'X_pca' not in adata.obsm.keys():
    sc.tl.pca(adata)

# Compute nearest neighbors
sc.pp.neighbors(adata, n_neighbors=15, use_rep='X_pca')  # Adjust 'n_neighbors' as needed

# Run UMAP
sc.tl.umap(adata, min_dist=0.3)

# Plot UMAP colored by cell type
sc.pl.umap(adata, color="cell_ontology", title="UMAP Colored by Cell Type")
```
