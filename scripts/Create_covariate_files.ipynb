```python
adata_combined = sc.read("PD_control_midbrain_aggregated_scobject.h5ad")
```

```python
adata_combined.obs
```

```python
#The covariates that we want are ["n_genes_by_counts", "Age at Death", "Sex", "pmi (h)"]
covariates = ["n_genes_by_counts", "Age at Death", "Sex", "pmi (h)"]
adata_combined_covars = adata_combined.obs[covariates]
```

```python
#Need to add a const with just 1 that is used to recentering estimates during the regression according to scDRS paper
adata_combined_covars["const"] = 1
#Encode the 'Sex' column: 1 for 'M' and 0 for 'F'
adata_combined_covars["Sex"] = adata_combined_covars["Sex"].map({"M": 1, "F": 0})
#Rename the Age at Death and pmi columns to replace the whitespace with underscore
adata_combined_covars.rename(columns={"Age at Death": "Age_at_Death", "pmi (h)": "pmi_h"}, inplace=True)
adata_combined_covars
```

```python
adata_combined_covars.index.is_unique
```

```python
#Export the covariate file
adata_combined_covars.to_csv("PD_control_midbrain_aggregated_covar_file.txt", sep="\t")
```
